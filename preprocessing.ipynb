```python
def woe_discrete(df, variable, target):
    df = pd.concat([df[variable], target], axis = 1)
    df = pd.concat([
        df.groupby(df.columns[0], as_index=False)[df.columns[1]].count(),
        df.groupby(df.columns[0], as_index=False)[df.columns[1]].mean(),  
    ], axis =1)
    df = df.iloc[:, [0,1,3]]
    df.columns = [variable,'n_obs','prop_bad']
    df['prop_good'] = 1 - df['prop_bad']
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_bad'] = df['prop_bad'] * df['n_obs']
    df['n_good'] = df['prop_good'] * df['n_obs']
    df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum()
    df['prop_n_good'] = df['n_good'] / df['n_good'].sum()
    df['WoE'] = np.log(df['prop_n_good'] / df['prop_n_bad'])
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
    df['IV'] = df['IV'].sum()
    df = df.sort_values('WoE').reset_index(drop=True)
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    return df
```

```python
def plot_by_woe(df_WoE, rotation= 0):
    x = np.array(df_WoE.iloc[: , 0].apply(str))
    y = df_WoE['WoE']
    plt.figure(figsize = (18, 6))
    plt.plot(x,y, marker ='o', linestyle = '--', color = 'k')
    plt.xlabel(df_WoE.columns[0])
    plt.ylabel('Weight of Evidence')
    plt.title(str('Weight of Evidence ' + df_WoE.columns[0]))
    plt.xticks(rotation = rotation)
```

## Continuous variables - WoE

```python
def woe_ordered_continuous(df, variable, target):
    df = pd.concat([df[variable], target], axis=1)
    df = pd.concat([
        df.groupby(df.columns[0], as_index=False)[df.columns[1]].count(),
        df.groupby(df.columns[0], as_index=False)[df.columns[1]].mean()
    ], axis=1)
    df = df.iloc[:, [0, 1, 3]]
    df.columns = [variable, 'n_obs', 'prop_bad']
    df['prop_good'] = 1 - df['prop_bad']
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_bad'] = df['prop_bad'] * df['n_obs']
    df['n_good'] = df['prop_good'] * df['n_obs']
    df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum()
    df['prop_n_good'] = df['n_good'] / df['n_good'].sum()
    df['WoE'] = np.log(df['prop_n_good'] / df['prop_n_bad'])
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
    df['IV'] = df['IV'].sum()
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    return df
```
